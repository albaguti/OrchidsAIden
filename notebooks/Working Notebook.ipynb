```python
import pandas as pd
import numpy as np
import imageio.v3 as iio
```

```python
orchids = pd.read_csv("data/orchids.csv")
orchids.head()
```

```python
genus = pd.read_csv("data/genus.csv")
genus.head()
```

We will create a directori for each of the genus that have pictures

```python
i ="Apostasia"
tmp = orchids[orchids.Genus == i]["href"]
tmp[2][2:-2].replace("',","").replace("'","").split()[1]
```

```python
gen = []
for i in genus.Genus:
    tmp = orchids[orchids.Genus == i]["href"][2:-2]
    for j in tmp:
        if j != "":
            tem = j.replace("',","").replace("'","").replace("[","").replace("]","").split()
            for k in tem:
                gen.append([i,k])
        
```

```python
len(gen)
```

```python
genus_with_image = []
image = []
for i in gen:
    if i[0] not in genus_with_image:
        genus_with_image.append(i[0])
        image.append(i[1])
```

```python
len(genus_with_image)
len(image)
```

```python
ge = pd.DataFrame(genus_with_image, columns=["Genus"])
```

```python
ge["href"] = image
```

```python
ge.to_csv("data/genus_with_image.csv", header=True)
```
