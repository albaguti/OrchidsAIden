```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import imageio.v3 as iio
```

<h1> Scrapping without Selenium

```python
base_url = "https://www.worldplants.de/world-orchids/orchid-list"
r = requests.get(base_url)
# print request object
print(r.url)

# print status code
print(r.status_code)
```

```python
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
 
s = soup.find('select', id='genus')
results = s.find_all('option')
genus = []
values = []
for result in results:
    genus.append(result.text)
    values.append(result.get('value'))

genus.remove("...")
values.remove("0")
```

We generate the url using the values

```python
# we generate the url
url = base_url + "#g-"+values[0]
browser = webdriver.Chrome()

browser.get(url)
wait = WebDriverWait(browser,20)
html = browser.page_source
```

<h2> Genus distribution

```python
# We get distribution information on the Genus
genus_info = browser.find_elements(By.CSS_SELECTOR, '.genusInfo')
dis = []
for e in genus_info:
    text = e.text
    #print(e.text)
    dis = text.split("\n")
genus_distribution = dis[1]
```

<h2> Species and Distribution

```python
orquid_selection = browser.find_elements(By.CSS_SELECTOR, ".type-species")
name = []
dist = []
href = []
h = []
for i in orquid_selection:
    n = i.find_element(By.CSS_SELECTOR, ".latName").text
    name.append(n)
    tmp = i.find_elements(By.CSS_SELECTOR, ".distCountry")
    for j in tmp:
        dist.append([n,j.text]) 
    image = i.find_elements(By.CSS_SELECTOR, "div.imageBox a")
    #print(image)
    for j in image:
        tmp =j.get_attribute("href")
        if tmp not in href:
            h.append(tmp)    
    href.append([n,h])
    h = []
```

```python
#Let's try an alternative method to retrieve the sub-species
orquid_selection = browser.find_elements(By.CSS_SELECTOR, ".type-species")
name = []
dist = []
href = []
h = []
data = []
distribution =[]
for i in orquid_selection:
    u_id =int(i.get_attribute("id").split("-")[1])
    n = i.find_element(By.CSS_SELECTOR, ".latName").text
    typ = "specie"
    name.append([u_id,n,typ])
    tmp = i.find_elements(By.CSS_SELECTOR, ".distCountry")
    for j in tmp:
        dist.append(j.text)
    distribution.append(dist)
    dist = []
    image = i.find_elements(By.CSS_SELECTOR, "div.imageBox a")
    #print(image)
    for j in image:
        tmp =j.get_attribute("href")
        if tmp not in h:
            h.append(tmp)    
    href.append([u_id,h])
    h = []
    data.append([u_id,n,distribution,h])
    
```

```python
orchid_sub = browser.find_elements(By.CSS_SELECTOR, ".type-subspecies")
name_sub = []
dist_sub = []
href_sub = []
h = []
data_sub = []
distribution = []
for i in orchid_sub:
    u_id =int(i.get_attribute("id").split("-")[1])
    n = i.find_element(By.CSS_SELECTOR, ".latName").text
    typ = "sub-specie"
    name_sub.append([u_id,n,typ])
    tmp = i.find_elements(By.CSS_SELECTOR, ".distCountry")
    for j in tmp:
        dist_sub.append([j.text])
    distribution.append(dist_sub)
    dist_sub = []
    image = i.find_elements(By.CSS_SELECTOR, "div.imageBox a")
    #print(image)
    for j in image:
        tmp =j.get_attribute("href")
        if tmp not in h:
            h.append(tmp)
    href_sub.append(h)
    data_sub.append([u_id,n,distribution,h])
    h = []
```

```python
data = pd.DataFrame(data, columns =["U-ID","Specie","Distribution","href"])
data["Sub-Specie"] = ""
data
```

```python
data_sub = pd.DataFrame(data_sub, columns =["U-ID","Sub-Specie","Distribution","href"])
data_sub["Specie"] = ""
data_sub
```

```python
j = 0
k = 0
for i in name_sub:
    while k == 0:
        if (i[0]-j) in data["U-ID"].values:
            spe = data[data["U-ID"] == (i[0]-j)]["Specie"].unique()
            data_sub.loc[data_sub["U-ID"]==i[0],"Specie"] = spe
            k = 1
        j += 1
    j = 0
    k = 0    
```

```python
data = pd.concat([data,data_sub],ignore_index = True)
data
```
