```python
genus = pd.read_csv("data/genus2.csv")
genus.head()
```

```python
base_url = "https://www.worldplants.de/world-orchids/orchid-list"
```

```python
def connection_to_webpage(base_url,value):
# we generate the url
    url = base_url + "#g-"+value
    browser = webdriver.Chrome()

    browser.get(url)
    wait = WebDriverWait(browser,20)
    html = browser.page_source

    return browser
```

```python
def get_genus(browser):
    # We get distribution information on the Genus
    sleep(2.5)
    genus_info = browser.find_elements(By.CSS_SELECTOR, '.genusInfo')
    dis = []
    distribution =""
    for e in genus_info:
        text = e.text
        dis = text.split("\n")
    for i in dis:
        if "Distribution:" in i:
            distribution = i
    return distribution
```

```python
data = pd.read_csv("data/genus.csv")
dis = []
for i in range(len(genus)):
    v = genus["Value"][i]
    g = genus["Genus"][i]
    if v not in data["Value"].values:
        url = base_url + "#g-"+ str(v)
        browser = webdriver.Chrome()
        browser.get(url)
        html = browser.page_source
        genus_distribution = get_genus(browser)
        print(genus_distribution)
        print(f"row {i}")
        dis.append([v,genus_distribution])  
        browser.close()
```

```python
dis = pd.DataFrame(dis, columns=["Value","Distribution"])
dis.head()
dis2 = dis.copy()
```
